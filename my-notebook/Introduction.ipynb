```python
# Write a function for saving yearly data files into a function. 
# First, write a function that separates data for just one year 
# and saves that data to a file. 

def one_year_csv_writer(this_year, all_data):
    """
    Writes a csv file for data from a given year.

    this_year --- year for which data is extracted
    all_data --- DataFrame with multi-year data
    """

    # Select data for the year
    surveys_year = all_data[all_data.year == this_year]

    # Write the new DataFrame to a csv file
    filename = 'data/yearly_files/function_surveys' + str(this_year) + '.csv'
    surveys_year.to_csv(filename)
```

```python
import pandas as pd
surveys_complete = pd.read_csv('data/surveys.csv')

# Remove the NA values
surveys_complete = surveys_complete.dropna()
```

```python
# Plotting with plotnine 

# To build a plotnine graphic we first need to bind the plot to a 
# specific DF using the data argument:

(p9.ggplot(data=surveys_complete))


# The output will be a empty figure because we have not defined
# anything else yet. 
```

```python
# Then, define aesthetics(aes) by selecting variables used 
# in the plot and mapping them to a presentation such as size, 
# shape, color, etc. 

# The most important aes mappings are: x, y, alpha, color, colour, 
# linetype, fill, shape, size, and stroke.

(p9.ggplot(data=surveys_complete, 
          mapping=p9.aes(x='weight', y='hindfoot_length')))
```

```python
# Next, we still have to define what kind of geometry will be 
# used. Use points( one of the geoms options), and the + operator
challenge 
surveys_plot = (p9.ggplot(data=surveys_complete,
          mapping=p9.aes(x='weight', y='hindfoot_length')))

#Draw the plot
surveys_plot + p9.geom_point()
```

```python
# Challenge - bar chart

# Use the plot-id column to create a. bar-plot that counts the 
# number of records for each plot. 

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='plot_id'))
    + p9.geom_bar()
)
```

```python
# Building your plots iteratively

# 1. define the dataset we use
# 2. lay the axes
# 3. choose a geom


(p9.ggplot(data=surveys_complete, 
          mapping=p9.aes(x='weight', y='hindfoot_length'))
    + p9.geom_point()
)
```

```python
# start modifying this plot to extract more info from it. Ex . add
# transparency (alpha) to avoid overplotting:

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
    + p9.geom_point(alpha=0.1)
)
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
    + p9.geom_point(alpha=0.1, color='blue')
)

# adds color for all the points
```

```python
# to color each species in the plot differently, map the 
# species_id column to the color aes:

(p9.ggplot(data=surveys_complete,
          mapping=p9.aes(x='weight', 
                        y='hindfoot_length',
                        color='species_id'))
    + p9.geom_point(alpha=0.1)
)
```

```python
# Defining scale for colors, axes
#Ex. a log version of the x-axis could make lower numbers easier
# to understand:

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
    + p9.geom_point(alpha=0.1)
    + p9.xlab("Weight (g)")
    + p9.scale_x_log10()
)
```

```python
# Change the theme, set the background to white 

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
    + p9.geom_point(alpha=0.1)
    + p9.xlab("Weight (g)")
    + p9.scale_x_log10()
    + p9.theme_bw()
    + p9.theme(text=p9.element_text(size=16))
)
```

```python
# Challenge - Bar plot adaptations

# Adapt the bar plot of the previous exercise by mapping the sex
# variable to the color fill of the bar chart. Change the scale of 
# the color fill by providing the colors blue and orange manually 

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='plot_id', 
                         fill='sex'))
    + p9.geom_bar()
    + p9.scale_fill_manual(["blue", "orange"])
)
```

```python
# Plotting distributions

# To visualize the distribution of weight within each species_id
# group, a boxplot can be used:

(p9.ggplot(data=surveys_complete,
          mapping=p9.aes(x='species_id', 
                        y='weight'))
    + p9.geom_boxplot()
)
```

```python
# We can add points of the individual observations to the boxplot

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='species_id',
                          y='weight'))
     + p9.geom_jitter(alpha=0.2)
     + p9.geom_boxplot(alpha=0.)
)
```

```python
# Challenge - distributions

# 1. Replace the box plot with a violin plot. 
# 2. Represent weight on the log10 scale
# 3. Add color fo the datapoints according to the plot 
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='species_id',
                          y='weight',
                          color='factor(plot_id)'))
    + p9.geom_jitter(alpha=0.3)
    + p9.geom_violin(alpha=0, color="0.7")
    + p9.scale_y_log10()
)
```

```python
# Plotting time series data 
# Calculate number of counts per year for each species. Group data
# first and count the species (species_id) within each group.

yearly_counts = surveys_complete.groupby(['year', 'species_id'])['species_id'].count()
yearly_counts
```

```python
# visualize as a line plot with years on x axis and counts on the
# y axis.

(p9.ggplot(data=yearly_counts,
          mapping=p9.aes(x='year', 
                        y='counts'))
     + p9.geom_line()
)
```

```python
# The above doesn't work. 
# Solution: draw a line for each species by modifying aes function
# and map the species_id to the color:

(p9.ggplot(data=yearly_counts,
           mapping=p9.aes(x='year',
                          y='counts',
                          color='species_id'))
    + p9.geom_line()
)
```

```python
# Faceting 

# Consider the scatter plot of weight vs hindfoot_length

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha=0.1)
)
```

```python
# split the graph and make a separate graph for each of the groups 
# in a chosen variable. Ex use 'sex'

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha=0.1)
    + p9.facet_wrap("sex")
)
```

```python
# Do the same for other variables, such as plot
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha=0.1)
    + p9.facet_wrap("plot_id")
)
```

```python
# You can specificy how you want your plots to be arranged:
# use 'rows ~ columns'

# only select the years of interest
survey_2000 = surveys_complete[surveys_complete["year"].isin([2000, 2001])]

(p9.ggplot(data=survey_2000,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha=0.1)
    + p9.facet_grid("year ~ sex")
)
```

```python
# Create a separate plot for each of the species that depicts how the
# average weight of the species changes through the years.

yearly_weight = surveys_complete.groupby(['year', 'species_id'])['weight'].mean().reset_index()

(p9.ggplot(data=yearly_weight,
           mapping=p9.aes(x='year',
                          y='weight'))
    + p9.geom_line()
    + p9.facet_wrap("species_id")
)
```

```python
# Visually compare how the weights of male and females has changed 
# through time by creating a. separate plot for each sex 
# and an individual color assigned to each species_id.

yearly_weight = surveys_complete.groupby(['year', 'species_id', 'sex'])['weight'].mean().reset_index()

(p9.ggplot(data=yearly_weight, 
          mapping = p9.aes(x='year',
                          y='weight',
                          color='species_id'))
    + p9.geom_line()
    + p9.facet_wrap("sex"))
```

```python
# Further customization 

# Bar graphs of count per year
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='factor(year)'))
    + p9.geom_bar()
)
```

```python
# To rotate the text of the x-axis labels

(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='factor(year)'))
    + p9.geom_bar()
    + p9.theme_bw()
    + p9.theme(axis_text_x = p9.element_text(angle=90))
)
```

```python
# save theme customizations as an object to easily apply them
# to other plots

my_custom_theme = p9.theme(axis_text_x = p9.element_text(color="grey", size=10,
                                                         angle=90, hjust=.5),
                           axis_text_y = p9.element_text(color="grey", size=10))
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='factor(year)'))
    + p9.geom_bar()
    + my_custom_theme
)
```
